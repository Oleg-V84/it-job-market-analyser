### Workplace column inspection

The `Workplace` column contains both English and French category names:

- `Hybrid` and `Hybride`
- `On-site` and `Sur cite`

The records with `Sur cite` are located in Québec and contain other French-language fields (`Développement Logiciel`, `Temps plein`), suggesting that these values are localized versions of existing workplace categories.

These categories will be standardized during the data cleaning stage.
